```python
import random
import pandas as pd

# 질병별 특정 염기서열 설정 (임의로 설정)
alzheimers_sequence = "ATGCAGTCA"  # 알츠하이머 질병을 위한 염기서열 (예시)
huntingtons_sequence = "GCTAGGCTAG"  # 헌팅턴병을 위한 염기서열 (예시)
sickle_cell_sequence = "TGCAGTGCAG"  # 낫모양 적혈구 빈혈증을 위한 염기서열 (예시)

# 염기서열 생성 함수
def generate_random_sequence(length=20):
    return ''.join(random.choice('ATGC') for _ in range(length))

# 50명의 염기서열 생성 (양성/음성 섞기)
data = []
for _ in range(50):
    # 각 질병에 대한 양성/음성 판별
    alzheimer_pos = random.choice([True, False])
    huntington_pos = random.choice([True, False])
    sickle_cell_pos = random.choice([True, False])
    
    # 염기서열 생성
    sequence = generate_random_sequence()
    
    # 양성 판별: 질병의 특정 염기서열 포함 여부로 판별
    if alzheimer_pos:
        sequence = sequence[:5] + alzheimers_sequence + sequence[15:]  # 알츠하이머 염기서열 삽입
    if huntington_pos:
        sequence = sequence[:5] + huntingtons_sequence + sequence[15:]  # 헌팅턴병 염기서열 삽입
    if sickle_cell_pos:
        sequence = sequence[:5] + sickle_cell_sequence + sequence[15:]  # 낫모양 적혈구 빈혈증 염기서열 삽입
    
    # 데이터 추가
    data.append({
        'ID': _+1,
        'Sequence': sequence,
        'Alzheimers_Positive': alzheimer_pos,
        'Huntingtons_Positive': huntington_pos,
        'Sickle_Cell_Positive': sickle_cell_pos
    })

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# CSV 파일로 저장
df.to_csv('genetic_data.csv', index=False)

# 결과 확인
print("CSV 파일이 'genetic_data.csv'로 저장되었습니다.")
```

```python
import pandas as pd

# 질병별 특정 염기서열 설정 (임의로 설정)
alzheimers_sequence = "ATGCAGTCA"  # 알츠하이머 질병을 위한 염기서열 (예시)
huntingtons_sequence = "GCTAGGCTAG"  # 헌팅턴병을 위한 염기서열 (예시)
sickle_cell_sequence = "TGCAGTGCAG"  # 낫모양 적혈구 빈혈증을 위한 염기서열 (예시)

# CSV 파일 읽기
df = pd.read_csv('genetic_data.csv')

# 각 염기서열에 대한 양성/음성 여부 판단
def check_disease(sequence, disease_sequence):
    """특정 질병에 대한 염기서열 포함 여부를 확인하는 함수"""
    if disease_sequence in sequence:
        return True  # 질병 염기서열이 포함되었을 경우 양성
    return False  # 질병 염기서열이 포함되지 않았을 경우 음성

# 각 질병에 대해 양성/음성 판단
df['Alzheimers_Prediction'] = df['Sequence'].apply(lambda x: check_disease(x, alzheimers_sequence))
df['Huntingtons_Prediction'] = df['Sequence'].apply(lambda x: check_disease(x, huntingtons_sequence))
df['Sickle_Cell_Prediction'] = df['Sequence'].apply(lambda x: check_disease(x, sickle_cell_sequence))

# 결과 출력
print(df[['ID', 'Sequence', 'Alzheimers_Prediction', 'Huntingtons_Prediction', 'Sickle_Cell_Prediction']])

# 결과를 새로운 CSV 파일로 저장 (예: 예측 결과 포함)
df.to_csv('genetic_data_with_predictions.csv', index=False)
print("결과를 'genetic_data_with_predictions.csv' 파일로 저장했습니다.")
```
