```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1.    
np.random.seed(42)
num_farms = 10
x = np.random.uniform(0, 100, size=num_farms)   #  
y = np.random.uniform(0, 100, size=num_farms)   #  
temperature = np.random.uniform(15, 35, size=num_farms)  # 
rainfall = np.random.uniform(50, 200, size=num_farms)    # mm

# 2.  :    + noise
#  = 2* + 0.5* + noise
true_weights = np.array([2.0, 0.5])
features = np.column_stack((temperature, rainfall))
noise = np.random.normal(0, 10, size=num_farms)
yield_est = features @ true_weights + noise

# 3. DataFrame 
df = pd.DataFrame({
    'x': x,
    'y': y,
    'Temperature': temperature,
    'Rainfall': rainfall,
    'Yield': yield_est
})

# 4. 
plt.figure(figsize=(8, 6))
sc = plt.scatter(df['x'], df['y'], c=df['Yield'], cmap='YlGn', s=200, edgecolor='k')
plt.colorbar(sc, label='Estimated Yield')
plt.xlabel('Longitude-like (x)')
plt.ylabel('Latitude-like (y)')
plt.title('Simulated Farm Yields based on Environmental Conditions')

#   ()
for i in range(len(df)):
    plt.text(df['x'][i]+1, df['y'][i]+1, f"{df['Yield'][i]:.1f}", fontsize=9)

plt.grid(True)
plt.show()
```

```python
import folium
import numpy as np
import pandas as pd

# 1.      ( )
farms = [
    {"name": "  A", "lat": 35.238, "lon": 128.691, "crop": ""},
    {"name": "  B", "lat": 35.245, "lon": 128.676, "crop": ""},
    {"name": "  C", "lat": 35.233, "lon": 128.661, "crop": ""},
    {"name": "  D", "lat": 35.221, "lon": 128.673, "crop": ""},
    {"name": "  E", "lat": 35.225, "lon": 128.699, "crop": ""},
]

# 2.    
np.random.seed(42)
n = len(farms)

sunlight = np.random.uniform(5, 10, n)      #   (hr/day)
moisture = np.random.uniform(30, 70, n)     #   (%)
slope = np.random.uniform(10, 30, n)        #  (%)
altitude = np.random.uniform(100, 400, n)   #  (m)

#   (   )
crop_factor = {
    "": 5.0,
    "": 3.0,
    "": 7.0,
    "": 9.0,
    "": 6.0
}

# 3.    (  )
a1, a2, a3, a4 = 10, 0.5, 1.0, 0.1  # 
yield_pred = []
for i, farm in enumerate(farms):
    base = (
        a1 * sunlight[i]
        + a2 * moisture[i]
        - a3 * slope[i]
        - a4 * altitude[i]
        + crop_factor[farm["crop"]]
        + np.random.normal(0, 3)  # 
    )
    yield_pred.append(max(base, 0))  #  

# 4. DataFrame 
df = pd.DataFrame(farms)
df["Sunlight"] = sunlight
df["Moisture"] = moisture
df["Slope"] = slope
df["Altitude"] = altitude
df["Yield"] = yield_pred

# 5.  
m = folium.Map(location=[35.2285, 128.6811], zoom_start=12)

# 6.  
for _, row in df.iterrows():
    color = "green" if row["Yield"] > 50 else "orange" if row["Yield"] > 30 else "red"
    popup_text = (
        f"{row['name']} ({row['crop']})<br>"
        f" : {row['Sunlight']:.1f} h<br>"
        f" : {row['Moisture']:.1f}%<br>"
        f" : {row['Slope']:.1f}%<br>"
        f" : {row['Altitude']:.0f} m<br>"
        f" : {row['Yield']:.1f}"
    )
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=9,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.8,
        popup=folium.Popup(popup_text, max_width=300)
    ).add_to(m)

m
```
