```python
import csv

# 질병 관련 염기서열 정의
alzheimers_sequence = "GCTGAGGAGGAGAGGAGGA"  # 알츠하이머병
huntingtons_sequence = "CAGCAGCAG"  # 헌팅턴병
sickle_cell_sequence = "GAGGTG"  # 낫모양 적혈구 증후군

# 질병 여부를 판단하는 함수
def check_disease_status(sequence):
    """
    염기서열에서 질병이 있는지 확인하는 함수
    :param sequence: 검사할 염기서열
    :return: 양성/음성 결과
    """
    diseases = []

    # 질병 관련 염기서열이 포함되어 있는지 검사
    if alzheimers_sequence in sequence:
        diseases.append("Alzheimer's Disease")
    if huntingtons_sequence in sequence:
        diseases.append("Huntington's Disease")
    if sickle_cell_sequence in sequence:
        diseases.append("Sickle Cell Disease")

    if diseases:
        return f"Positive ({', '.join(diseases)})"  # 양성
    else:
        return "Negative"  # 음성

# CSV 파일 경로
csv_file_path = 'sequences_mixed.csv'

# CSV 파일 열기 및 데이터 읽기
with open(csv_file_path, mode='r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # 첫 번째 행은 헤더이므로 건너뜀

    # 각 사람에 대해 염기서열을 확인하고 양성/음성 결과 출력
    for row in csv_reader:
        name = row[0]  # 이름
        sequence = row[1]  # 염기서열
        disease_status = check_disease_status(sequence)
        print(f"{name}: {disease_status}")

```
